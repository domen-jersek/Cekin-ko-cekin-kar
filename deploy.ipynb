## Option A: Fly.io (free-ish, smoothest with our new CI)

What you’ll need:
- Fly.io account and CLI (or just use the GitHub Actions flow).
- A Fly app name (change `app` in fly.toml or let `fly launch` set one).
- Secrets: DISCORD_TOKEN, GUILD_ID, SPREADSHEET_ID, SHEET_NAME, CEKINCKI_ROLE_PREFIX, SYNC_INTERVAL_SECONDS, NOTIFY_* (optional), and GCP_CREDENTIALS_JSON.

Two ways to deploy:

### A1) Deploy via GitHub Actions (recommended)
- In your GitHub repo settings:
  - Add a repository secret `FLY_API_TOKEN` (create this from Fly: account settings -> tokens).
- Push to `master` (already wired).
  - The workflow fly-deploy.yml will run `flyctl deploy --remote-only` and publish the app.
- Set Fly secrets once (run these locally with Fly CLI or in a Codespace — the CI doesn’t set app secrets):
  - After logging in to Fly, set secrets on the app. For PowerShell, use single quotes for the JSON:
    - DISCORD_TOKEN, GUILD_ID, SPREADSHEET_ID, SHEET_NAME, CEKINCKI_ROLE_PREFIX (e.g., “Cekinčki:”), SYNC_INTERVAL_SECONDS, NOTIFY_*.
    - GCP_CREDENTIALS_JSON: paste your Service Account JSON.

Tip: You can also set secrets from the Fly web UI if using Launchpad.

### A2) Manual deploy from your machine
- Install Fly CLI and login (once).
- In repo root:
  - Initialize app if needed: fly launch --no-deploy
  - Set secrets:
    - DISCORD_TOKEN, GUILD_ID, SPREADSHEET_ID, SHEET_NAME, CEKINCKI_ROLE_PREFIX, SYNC_INTERVAL_SECONDS, NOTIFY_*.
    - GCP_CREDENTIALS_JSON (paste entire JSON).
  - Deploy:
    - fly deploy
  - Watch logs:
    - fly logs

Notes:
- We intentionally removed `http_service` in fly.toml as this bot doesn’t expose HTTP. That’s correct for a long‑running worker.
- Make sure your Google Sheet is shared with the service account email.
